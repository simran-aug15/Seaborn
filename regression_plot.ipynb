```python
#import
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Regplot , lmplot 
# In the simplest invocation,both functions draw a scatterplot of two variables X and Y then fit the regression model Y ~ X and plot the resulting regression line 
# and 95% confidence interval for that regression
```

```python
# import dataset
tips=sns.load_dataset('tips')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# Now using regplot 
sns.regplot(data=tips,x='total_bill',y='tip')    #hue parameter doesnot exist in this
```

```python
#now using lmplot
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot : It gives the error between real value and given value
sns.residplot(data=tips,x='total_bill',y='tip')
```
