```python
#import
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# import database
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
# 1 Categorical scatter plot : a.Stripplot   b.swarmplot
# 2 Categorical Distribution plot : a.Boxplot  b.violinplot
# 3 Categorical Estimate plot : a. Barplot  b.pointplot  c.countplot
```

#### Categorical Scatter plot

```python
# Figure level function : catplot
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# strip plot : It takes one categorical data and other numerical data and then plot graph between
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
# axes level : swarmplot : Use an algorithm (good for small data)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

#### Categorical Distribution plots

```python
# Box plot:
# A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). 
# It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.
```

```python
# Box plot : axes level
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
# using catplot : figure level function
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
# single boxplot : numerical col
sns.boxplot (data=tips,y='total_bill')
```

```python
# Violinplot : (Boxplot + kdeplot)
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# catplot 
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

#### Categorical estimate plot

```python
# Barplot : When there are multiple observations in each category, it also uses bootstrapping to compute
# a confidence interval around the estimate, which is plotted using error bars.
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.std)
```
