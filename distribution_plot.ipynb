```python
#import
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Plots under this :
#1.histplot
#2.kdeplot
#3.rugplot
```

```python
# plotting univariate histogram
# figure level -> displot
# axes level ->histplot ->kdeplot ->rugplot
```

```python
tip=sns.load_dataset('tips')
tip
```

```python
# axes plot
sns.histplot(data=tip,x='total_bill')
```

```python
# figure plot
sns.displot(data=tip,x='total_bill',kind='hist')
```

```python
# bins parameter
sns.displot(data=tip,x='total_bill',kind='hist',bins=2)
```

```python
# It's also possible to visualize the distribution of the categorical variable using the logic of the histogram
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=tip,x='day',kind='hist')
```

```python
# hue parameter
sns.displot(data=tip,x='day',kind='hist',hue='sex')
```

```python
# element : step(give you outline not full bars)
sns.displot(data=tip,x='day',kind='hist',hue='sex',element='step')
```

```python
# Example 2.
titanic=sns.load_dataset('titanic')
titanic
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
plt.show()
```

```python
# Faceting using row and col
sns.displot(data=tip,x='tip',kind='hist',col='sex',element='step')
```

```python
# kdeplot : Rather than using discrete bins,a KDE plot smooths the observations with a Gaussian kernal,producing a continous density estimate
# axes level function 
# In this y axis represent density on probability of data
sns.kdeplot(data=tip,x='total_bill')
```
