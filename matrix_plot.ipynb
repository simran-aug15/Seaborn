```python
#import
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Heatmap : Plot rectanguler data as a color-encoded matrix
gap=px.data.gapminder()
gap
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
# axes level function(and it is only way to generate it)
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

```python
# annot : It gives the number
temp_df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(10,5))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='summer') 
# linewidth : Make space between boxes
```

```python
# clustermap : Same as heatmap ,but it make similar column together and dis-similar together
# Plot a matrix dataset as a hierarchially -clustered heatmap
# This function requires scipy to be available
iris=px.data.iris()
iris
```
