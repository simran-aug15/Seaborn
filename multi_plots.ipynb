```python
#import
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# FacetGrid : A second way to plot facet
# figure level : relplot->displot->catplot->lmplot
# First way
tips=sns.load_dataset('tips')
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
# other way
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid vs Pairplot)

```python
iris=sns.load_dataset('iris')
```

```python
# pairplot -> plot scatterplot between pairs of all the numerical columns in data
# when it will compare two same data -> it will plot  histogram
sns.pairplot(iris)
```

```python
# Pairplot : Detect how many numerical columns in data,and create scatterplot between each of them and between same values
sns.pairplot(iris,hue='species')
```

```python
# PairGrid
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
# map-diag ->map_offdiagonal This will create any graph as your wish in the diagonal and other on the non diagonal parts 
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot) 
```

```python
# map_diag -> map_upper ->map_lower 
# This will form a diagonal graph as histogram the graph above the diagonal will be kdeplot and lower one will be scatterplot 
# you can change the type according to ypur wish 
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars : Restrict the data on which you want to work
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.add_legend()
```
