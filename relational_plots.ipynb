#### Scatter plot :

```python
#import
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# axis level function
tip=sns.load_dataset('tips')
tip
sns.scatterplot(data=tip,x="total_bill",y="tip",hue='sex',style='time',size='size')
```

```python
# Figure level function : 
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',hue='sex',size='size',style='time')
```

```python
# lineplot :
gap=px.data.gapminder()
gap
temp_df=gap[gap['country']=='India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# hue-> style
temp_df=gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
# relplot : Figure level function
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
# To show that it will give continental info:
temp_df=gap[gap['country'].isin(['India','Europe','Germany'])]
temp_df
```

```python
# Lineplot : axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')# size= thickness and thinness of line
```

```python
# Facet plot : On the particular column you can create multiple columns.
# always on figyre level function 
# using one function for row
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',row='sex')
```

```python
# using function for both row and col
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',row='day',col='smoker')
```
