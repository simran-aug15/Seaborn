#### Scatter plot :

```python
#import
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# axis level function
tip=sns.load_dataset('tips')
tip
sns.scatterplot(data=tip,x="total_bill",y="tip",hue='sex',style='time',size='size')
```

```python
# Figure level function : 
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',hue='sex',size='size',style='time')
```

```python
# lineplot :
gap=px.data.gapminder()
gap
temp_df=gap[gap['country']=='India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```
